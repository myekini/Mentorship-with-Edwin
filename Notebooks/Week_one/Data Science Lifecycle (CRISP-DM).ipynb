# Data Science Lifecycle

### 1 - Business Understanding
  - 1. Understand the problem (A banks wants to predict customer churn to reduce revenue loss)
  - 2. Define the success metrics (reduce customer churn by atleast 30%-- deliver business value)

2 - Data Understanding
  - 1. Collection (diff data sources---- Database, CSV, Datawarehouse, Data-lake)
  - 2. description (Initial analysis --- columns, mean, median, standard deviation)
  - 3. exploration (EDA--- data cleaning, data transformation, data distribution, correlation analysis, data sampling)
  - 4. Data quality (Data has the right data types, standardized)

3 - Data Preparation 
  - 1. clean
  - 2. transform (enriching the dataset)
  - 3. model 

4 - Data Modeling (Train a ML algorithm)
  - 1. training the model (regression, classification , etc)
     - Supervised Learning algorithm (Decision tree, linear regression, logistic regression)
     - Unsupervised Learning algorithm (Classification (K-MEANS), LSTM, Neural networks)

  - 2. Evaluation (checks how well the model performed)
      - Accuracy of the model
      - Hyper-parameter fine-tuning
  - 3. Deployment (deliver to end users (AKA customers))
      - API development 
      - CI/CD
      - Cloud-based hosting (Server, Database hosted)
